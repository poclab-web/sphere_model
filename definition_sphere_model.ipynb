```python
param=[1.5,2,100]
center=[param[1]*np.cos(np.radians(param[2])),0,param[1]*np.sin(np.radians(param[2]))]
radius=param[0]
dir_name="/Volumes/SSD-PSM960U3-UW/b3lyp_6-31g(d)_b3lyp_6-31g(d)_single_point_calculation/KPMKEVXVVHNIEY-NTSWFWBYSA-N"#"/YKFKEYKJGVSEIX-BAUXJJEQSA-N"#"/KPMKEVXVVHNIEY-NTSWFWBYSA-N"


mol = py3Dmol.view({"viewergrid": None, "style": (10, 10)}).rotate(-90, "x")#.rotate(-90, "x")


with open(dir_name+"/optimized0.xyz", "r") as f:
    rl = f.read()
mol.addModel(rl,  'xyz', {"style": "stick"})
color="red"
mol_=copy.deepcopy(mol)
mol_.addModel(rl,  'xyz', {"style": "sphere"})
center_local=[param[1]*np.cos(np.radians(90)),0,param[1]*np.sin(np.radians(90))]
d=4
radius_=0.025
color_="gray"

# mol_.addArrow(
#     {"start": {"x": 0, "y": 0, "z": -d},
#      "end": {"x": 0, "y": 0, "z": d}, "radius": radius_,
#      "radiusRatio": 4, "mid": 0.9, "color": color_})
for i in range(200):
    k=(i/100)
    j=((i+1)/100)
    opacity=1-1*(i/200)
    radius_local=radius*center[2]*k/4
    mol_.addCylinder({"start":{"x": center_local[0]*k, "y": center_local[1]*k, "z": center_local[2]*k},
                     "end":{"x": center_local[0]*j, "y": center_local[1]*j, "z": center_local[2]*j},
                     "radius":radius_local,"color":color,'opacity':opacity,"wireframe": False})
    mol_.addCylinder({"start":{"x": center_local[0]*k, "y": center_local[1]*k, "z": -center_local[2]*k},
                     "end":{"x": center_local[0]*j, "y": center_local[1]*j, "z": -center_local[2]*j},
                     "radius":radius_local,"color":color,'opacity':opacity,"wireframe": False})
mol_.show()

###
mol_=copy.deepcopy(mol)
# render_moleculer.addxyzarrow(mol_)
for i in range(100):
    k=(1+i/100)
    j=(1+(i+1)/100)
    opacity=1-1*(i/100)
    mol_.addCylinder({"start":{"x": center[0]*k, "y": center[1]*k, "z": center[2]*k},
                     "end":{"x": center[0]*j, "y": center[1]*j, "z": center[2]*j},
                     "radius":radius,"color":color,'opacity':opacity,"wireframe": False})
    mol_.addCylinder({"start":{"x": center[0]*k, "y": center[1]*k, "z": -center[2]*k},
                     "end":{"x": center[0]*j, "y": center[1]*j, "z": -center[2]*j},
                     "radius":radius,"color":color,'opacity':opacity,"wireframe": False})
mol_.addSphere(
        {"center": {"x": center[0], "y": center[1], "z": center[2]}, 'opacity': 1, "radius": radius,
         "color": color,"wireframe":False})
mol_.addSphere(
        {"center": {"x": center[0], "y": center[1], "z": -center[2]}, 'opacity': 1, "radius": radius,
         "color": color,"wireframe":False})
mol_.show()

###
if False:
    with open(dir_name+"/Dt0.cube", "r") as f:
        rl = f.read()
    render_moleculer.add_label(mol,param,center)
    mol.addSphere(
            {"center": {"x": center[0], "y": center[1], "z": center[2]}, 'opacity': 0.9, "radius": radius,
             "color": color,"wireframe":True})
    
    
    mol.addSphere(
            {"center": {"x": center[0], "y": center[1], "z": -center[2]}, 'opacity': 0.9, "radius": radius,
             "color": color,"wireframe":True})
    mol_=copy.deepcopy(mol)
    
    #mol_.addVolumetricData(rl, 'cube', {'isoval': 0.02, 'color': 'black', 'opacity': 0.5, "wireframe": True})
    for isoval,opacity in zip([0.02],[1]):
        mol_.addVolumetricData(rl, 'cube', {'isoval': isoval, 'color': 'orange', 'opacity': opacity, "wireframe": True})
    mol_.show()
    
    with open(dir_name+"/LUMO0.cube", "r") as f:
        rl = f.read()
    mol_=copy.deepcopy(mol)
    for isoval,opacity in zip([0.02],[0.6]):
        mol_.addVolumetricData(rl,'cube', {'isoval': isoval, 'color': 'red', 'opacity': opacity, "wireframe": False})\
        .addVolumetricData(rl, 'cube', {'isoval': -isoval, 'color': 'blue', 'opacity': opacity, "wireframe": False})
    mol_.show()
```
